```python
import pandas as pd
import numpy as np
import nltk
import pathlib
from nltk.tokenize import word_tokenize
```

```python
text_dir = pathlib.Path('bbc-fulltext/bbc/business/')

sentences = []
for text_path in text_dir.glob("*.txt"):
    s = text_path.read_text()
    sentences.extend(word_tokenize(s))
```

```python
sentences
```

```python
# text = pathlib.Path('bbc-fulltext/bbc/business/001.txt').read_text()
# for i in text:
#     print(i)
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
stop_words = set(stopwords.words('english'))
filtered_sentence = [w for w in sentences if not w.lower() in stop_words]
filtered_sentence
```

```python
from nltk.stem import PorterStemmer,LancasterStemmer
pstemmer = PorterStemmer()
stemp = []
for w in filtered_sentence:
    h = pstemmer.stem(w)
    stemp.append(h)
stemp
```

```python
lstemmer = LancasterStemmer()
ltemp = []
for w in filtered_sentence:
    h = pstemmer.stem(w)
    ltemp.append(h)
ltemp
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemma = WordNetLemmatizer()
lemmas = []
for w in filtered_sentence:
    h = lemma.lemmatize(w)
    lemmas.append(h)
lemmas
```

```python
# nltk.download('omw-1.4')
```

```python
vocabulary = {}
for word in lemmas:
    if word in vocabulary:
        vocabulary[word] += 1
    else:
        vocabulary[word] = 1
```

```python
vocabulary
```

```python
with open("business.txt", "w") as f:
    for word, count in vocabulary.items():
        f.write("{}: {}\n".format(word, count))
```

```python
with open("business.txt","r") as f:
    for line in f:
        print(line)
```
